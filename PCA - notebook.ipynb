```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
data = np.genfromtxt('/home/sebastian/Descargas/pca_2d/pcaData.txt', dtype='float32', delimiter="  ")
```

```python
len(data[0]) == len(data[1])
```

```python
plt.scatter(data[0],data[1])
plt.show()
```

```python
sigma = np.dot(data, data.transpose())  / data.shape[1]
(u, s, v) = np.linalg.svd(sigma)
```

```python
u
```

```python
plt.scatter(data[0],data[1])
plt.plot([0, u[0][0]], [0, u[0][1]], '-')
plt.plot([0, u[1][0]], [0, u[1][1]], '-')
plt.show()
```

```python
xRot = np.dot(u.transpose(), data)
xRot
```

```python
plt.scatter(xRot[0],xRot[1])
plt.show()
```

```python
k = 1
xHat = np.dot(u[:,:k].transpose(), data)
```

```python
plt.scatter(xHat[0],xHat[0])
plt.show()
```

Whitening

```python
epsilon = 1e-5
```

```python
s
```

```python
xPCAwhite = np.dot(np.dot(np.diag(1./np.sqrt(s + epsilon)), u.transpose()), data)
# This is the same to  xPCAwhite = np.dot(np.dot(np.diag(1./np.sqrt(s + epsilon)), xRot)
```

```python
plt.scatter(xPCAwhite[0],xPCAwhite[1])
plt.show()
```

ZCA Whitening

```python
xZCAwhite = np.dot(u, np.dot(np.dot(np.diag(1./np.sqrt(s + epsilon)), u.transpose()), data))
#This is the same to xZCAwhite = np.dot(u, xPCAwhite)
```

```python
plt.scatter(xZCAwhite[0],xZCAwhite[1])
plt.show()
```
